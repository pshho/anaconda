### 연령별 인구 시각화

```python
import csv
import matplotlib.pyplot as plt

with open('./datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    # next(data)
    result = list() # 연령별 인구 저장 리스트

    for row in data:
        if '신림' in row[0]:
            # print(row)
            for i in row[3:]: # 3번 인덱스부터 ~ 104번 인덱스까지 반복
                result.append(int(i.replace(',', ''))) # 천단위 구분기호 제거

# print(len(result))
# print(result)

plt.rc('font', family='Malgun Gothic')
plt.style.use("ggplot") # 격자 무늬 스타일
plt.title('연령별 인구 통계')
plt.plot(result[:101]) # result - y축값
plt.xlabel('연령')
plt.ylabel('인구')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

with open('./datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    # next(data)
    result = list() # 연령별 인구 저장 리스트
    name = input("지역의 이름을 입력하세요\n")

    for row in data:
        if name in row[0]:
            # print(row)
            for i in row[3:]: # 3번 인덱스부터 ~ 104번 인덱스까지 반복
                result.append(int(i.replace(',', ''))) # 천단위 구분기호 제거

# print(len(result))
# print(result)

plt.rc('font', family='Malgun Gothic')
plt.style.use("ggplot") # 격자 무늬 스타일
plt.title(f'연령별 인구 통계({name})')
# plt.plot(result[:101]) # result - y축값
plt.bar(range(101), result[:101]) # 막대그래프: x축(0-100), y축(result)
plt.xlabel('연령')
plt.ylabel('인구')
plt.show()
```
